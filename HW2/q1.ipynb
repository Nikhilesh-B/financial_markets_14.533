```python
def subtract_months(date_str, months_to_subtract):
    """
    Subtract months from a date string in YYYYMM format
    
    Args:
        date_str (str): Date in YYYYMM format (e.g., '192712')
        months_to_subtract (int): Number of months to subtract
    
    Returns:
        str: New date in YYYYMM format
    """
    year = int(date_str[:4])
    month = int(date_str[4:])
    
    # Convert to total months
    total_months = year * 12 + month - 1  # -1 for 0-indexing
    
    # Subtract months
    total_months -= months_to_subtract
    
    # Convert back to year and month
    new_year = total_months // 12
    new_month = (total_months % 12) + 1  # +1 to get back to 1-indexing
    
    # Format as YYYYMM string
    return f"{new_year:04d}{new_month:02d}"

# Usage examples:
date = '192712'
print(subtract_months(date, 1))   # Output: '192711'
print(subtract_months(date, 12))  # Output: '192612'
print(subtract_months(date, 24))  # Output: '192512'
print(subtract_months(date, 13))  # Output: '192611'
```
